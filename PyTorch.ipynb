# **Tensors in Python**

# Creating a Tensor

```python
import torch
print(torch.__version__)
```

```python
if torch.cuda.is_available():
    print("GPU is available!")
    print(f"Using GPU: {torch.cuda.get_device_name(0)}")
else:
    print("GPU not available. Using CPU.")
```

```python
# using empty
a = torch.empty(2,3)
a
```

```python
# check type
type(a)
```

```python
# using zeros
torch.zeros(2,3)
```

```python
# using ones
torch.ones(2,3)
```

```python
# using rand
torch.rand(2,3)
```

```python
# manual_seed  ## Seed is used to generate same random numbers in each execution
torch.manual_seed(100)
torch.rand(2,3)
```

```python
torch.manual_seed(100)
torch.rand(2,3)
```

```python
# using tensor
torch.tensor([[1,2,3],[4,5,6]])
```

```python
# # using tensor
# torch.tensor([[1,2,4],[4,5]])  ## error
```

```python
# other ways

# arange
print("using arange ->", torch.arange(0,10,2))

# using linspace
print("using linspace ->", torch.linspace(0,10,10))

# using eye
print("using eye ->", torch.eye(5))

# using full
print("using full ->", torch.full((3, 3), 5))
print("using full ->", torch.full((3, 2), 5))
```

# Tensor Shapes

```python
x = torch.tensor([[1,2,3],[4,5,6]])
x
```

```python
x.shape
```

```python
torch.empty_like(x)  ## create an empty  tensor with dimension similar as x
```

```python
torch.zeros_like(x)  ## create a zero tensor with dimension similar as x
```

```python
torch.rand_like(x, dtype=torch.float32)   ## As random function stores float value, you have to mention it here.
```

# Tensor Data Types

```python
# find data type
x.dtype
```

```python
# assign data type
torch.tensor([1.0,2.0,3.0], dtype=torch.int32)
```

```python
torch.tensor([1,2,3], dtype=torch.float64)
```

```python
# using to()
x.to(torch.float32)
```
